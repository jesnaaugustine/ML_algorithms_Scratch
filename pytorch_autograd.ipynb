```python
import torch
from torch.autograd import grad
import torch.nn.functional as F
```

```python
x = torch.tensor([3.,])
w = torch.tensor([2.,],requires_grad=True)
b = torch.tensor([1.,],requires_grad=True)
out = F.relu(torch.matmul(x,w)+b)
```

```python

grad(out,w,retain_graph=True)
```

```python
grad(out,b)
```

```python
grad(out,w)
```
