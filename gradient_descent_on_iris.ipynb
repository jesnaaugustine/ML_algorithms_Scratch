```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/Users/libinvarghese/Documents/Jesna/Git_ML/ML_algorithms_Scratch/data/Advertising.csv')
data.head()
```

```python
data.shape
```

```python
#features and target varible
X = data[['TV','Radio','Newspaper']]
y = data.Sales
```

```python
from Gradient_Descent import GradientDescent as GD

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
gd1 =GD(eta =0.01,n_iter=15,random_state=10)
gd1.fit(X,y)
ax[0].plot(range(1, len(gd1.losses) + 1),np.log10(gd1.losses), marker='o')
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('log(Mean squared error)')
ax[0].set_title('GD rate 0.1')
gd2 =GD(eta =0.000001,n_iter=15,random_state=10)
gd2.fit(X,y)
ax[1].plot(range(1, len(gd2.losses) + 1),np.log10(gd2.losses), marker='o')
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('Mean squared error')
ax[1].set_title('GD rate 0.0001')
    
plt.show()




```
