```python
import pandas as pd
import numpy as np
df_wine=pd.read_csv('/Users/libinvarghese/Documents/Jesna/Git_ML/ML_algorithms_Scratch/data/wine'
'.csv')
```

```python
#train_test_split
from sklearn.model_selection import train_test_split

X = df_wine.iloc[:,1:]
y = df_wine.iloc[:,1]
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
print(f'x_train:{x_train.shape} x_x_test:{x_test.shape}')
print(f'y_train:{y_train.shape} y_test:{y_test.shape}')
```

```python
#standrdize input data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaler = scaler.fit_transform(x_train)
x_test_scaler = scaler.transform(x_test)
```

```python
#covariance 
cov_data = np.cov(x_train_scaler.T)
eigen_vals, eigen_vecs = np.linalg.eig(cov_data)
eigen_vals.shape[0]
```

```python
eigen_pairs = [(np.abs(eigen_vals[i]), eigen_vecs[:,i]) for i in range(eigen_vals.shape[0])]
eigen_pairs.sort(key=lambda k:k[0],reverse=True)
```

```python
eigen_pairs
```

```python
W = np.empty((eigen_vecs.shape[1],4))
for i in range(4):
    W[:,i]=eigen_pairs[i][1]
print(W)
```

```python
x_train_pca = x_train_scaler.dot(W)
```
