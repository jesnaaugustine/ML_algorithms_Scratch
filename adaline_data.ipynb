```python
import torch
import numpy as np
from adaline import ADALINE
```

```python
X = torch.tensor([
    [2, 3],
    [1, 1],
    [2, 1],
    [-1, -1],
    [-2, -3],
    [-1, -2]
])

y = torch.tensor([1, 1, 1, -1, -1, -1])
model = ADALINE(lerrate=0.01)
model.fit(X,y)
```

```python
x_pre =torch.tensor([[2,3],[2,1],[-1, -2]])
model.predict(x_pre)
```
