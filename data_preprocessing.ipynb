```python
data = {
    'Department': ['IT', 'IT', 'HR', 'HR', 'IT', 'HR', 'IT', 'HR'],
    'Salary': [50000, 60000, 45000, np.nan, np.nan, 55000, 65000, np.nan],
    'Absent':['N','N','Y',np.nan,np.nan,'N','N','Y']
}
df_1 = pd.DataFrame(data)
```

```python
df_1.head()
```

```python
df_1['Salary'].fillna(df_1.groupby('Department')['Salary'].transform('mean').round(2),inplace=True)
```

```python
df_1
```

```python
df_1['Absent']=df_1['Absent'].fillna(df_1['Absent'].mode()[0])
```
