```python

import numpy as np
from perceptron import Perceptron,Perceptron_Pytorch
import torch
```

```python
X = np.array([
    [2, 3],
    [1, 1],
    [2, 1],
    [-1, -1],
    [-2, -3],
    [-1, -2]
])

y = np.array([1, 1, 1, -1, -1, -1])
model = Perceptron(learn_rate=0.01)
model.fit(X,y)
```

```python
model.predict(np.array([[2,3],[2,1],[-1, -2]]))
```

```python
X = torch.tensor([
    [2, 3],
    [1, 1],
    [2, 1],
    [-1, -1],
    [-2, -3],
    [-1, -2]
])

y = torch.tensor([1, 1, 1, -1, -1, -1])
model = Perceptron_Pytorch(learn_rate=0.01)
model.fit(X,y)
```

```python
x_pre =torch.tensor([[2,3],[2,1],[-1, -2]])
model.predict(x_pre)
```
