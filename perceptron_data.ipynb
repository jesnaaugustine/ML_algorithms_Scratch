```python
import numpy as np
from perceptron import Perceptron
```

```python
X = np.array([
    [2, 3],
    [1, 1],
    [2, 1],
    [-1, -1],
    [-2, -3],
    [-1, -2]
])

y = np.array([1, 1, 1, -1, -1, -1])
model = Perceptron(learn_rate=0.01)
model.fit(X,y)
```

```python
model.predict(np.array([[2,3],[2,1],[-1, -2]]))
```
